### 3.3 Datos Nulos y Tratamiento

Outliers: se tratarán sólo los valores negativos, y en age, se hará un tratamiento especial de los extremos, respecto de las demás, se utilizará un RobustScaler para tratarlos más adelante, ya que en son valores que podrían darse.

Principales limpiezas realizadas:
- age: se asume que los valores negativos son errores de tipeo, por lo que se los convierte a positivos, además, se asume que los valores extremos son errores de tipeo, por lo que se obtiene el residuo de la división por 100, para obtener valores mayores que 0 pero menores a 100.
- variables con valores atípicos negativos: al igual que con la edad, se asume que los valores negativos son errores de tipeo, por lo que se los convierte a positivos, como se mencionó anteriormente, los valores atípicos positivos o extremos, se corregirán con el escalador.
- ocupation: '_______' se asume como NoInfo.
- payment_behaviour: '!@9#%8' también se asume como NoInfo, esto para no perder observaciones.

Cabe mencionar que estas dos últimas, se requiere poder confirmar con mayor conocimiento experto de la generación de las bases de datos, pero como un acercamiento en primera instancia, son supuestos que pueden ser considerados razonables.
- Nulos: se utiliza un SimpleImputer con la media como estrategia.

```python
from sklearn.impute import SimpleImputer
def data_cleaner(X):
    # Outliers: se tratarán sólo los valores negativos, y en age, se hará un tratamiento especial de los extremos, respecto de las demás, se utilizará 
    # un RobustScaler para tratarlos más adelante, ya que en son valores que podrían darse.

    # age: valores negativos y extremos
    # se asume que los valores negativos son errores de tipeo, por lo que se los convierte a positivos
    X.loc[:,'age'] = X.loc[:,'age'].apply(lambda x: -x if x<0 else x)
    # se asume que los valores extremos son errores de tipeo, por lo que se obtiene el residuo de la división por 100
    X.loc[:,'age'] = X.loc[:,'age'].apply(lambda x: x%100 if x>100 else x)
    # se vuelve a valor entero
    X.loc[:,'age'] = pd.to_numeric(X.loc[:,'age'], errors='coerce', downcast='integer')
    
    # negativos
    negative = ['num_bank_accounts','num_of_loan','delay_from_due_date']
    for col in negative:
        X.loc[:,col] = X.loc[:,col].apply(lambda x: -x if x<0 else x)
        X.loc[:,col] = pd.to_numeric(X.loc[:,col], errors='coerce', downcast='integer')

    # occupation = '_______'
    X.loc[:,'occupation'] = X.loc[:,'occupation'].apply(lambda x: 'NoInfo' if x=='_______' else x)

    # payment_behaviour = '!@9#%8' 
    X.loc[:,'payment_behaviour'] = X.loc[:,'payment_behaviour'].apply(lambda x: 'NoInfo' if x=='!@9#%8' else x)


    # Nulos:
    nulls = ['changed_credit_limit','num_credit_inquiries','amount_invested_monthly','monthly_balance']
    imputer = SimpleImputer(strategy='mean')
    X[nulls] = imputer.fit_transform(X[nulls])
    return X
```

### 3.4 Feature Engineering (Bonus)

Se crean nuevas variables:

Ratios: 
- debt_to_income_ratio: ratio entre outstanding_debt y annual_income, es decir, cuánto porcentaje del ingreso anual es destinado a pagar deudas impagas.
- credit_card_utilization: ratio ente num_credit_cart y num_bank_accounts, es decir, cuántas tarjetas de crédito hay por cuentas de banco que tiene el cliente.

Discretización de age: en 'young', 'adult', 'middle-aged' y 'senior'.

```python
def Feature_Extractor(X):
    # Crear nuevas características
    X['debt_to_income_ratio'] = X['outstanding_debt'] / X['annual_income']
    X.loc[X['annual_income'] == 0, 'debt_to_income_ratio'] = 0  # Asignar 0 a los casos donde el denominador es 0

    X['credit_card_utilization'] = X['num_credit_card'] / X['num_bank_accounts']
    X.loc[X['num_bank_accounts'] == 0, 'credit_card_utilization'] = 0  # Asignar 0 a los casos donde el denominador es 0

     # age: valores negativos y extremos
    # se asume que los valores negativos son errores de tipeo, por lo que se los convierte a positivos
    X.loc[:,'age'] = X.loc[:,'age'].apply(lambda x: -x if x<0 else x)
    # se asume que los valores extremos son errores de tipeo, por lo que se obtiene el residuo de la división por 100
    X.loc[:,'age'] = X.loc[:,'age'].apply(lambda x: x%100 if x>100 else x)
    # se vuelve a valor entero
    X.loc[:,'age'] = pd.to_numeric(X.loc[:,'age'], errors='coerce', downcast='integer')

    
    # Discretizar la variable 'age'
    X['age_group'] = pd.cut(X['age'], bins=[0, 30, 40, 50, float('inf')],
                            labels=['young', 'adult', 'middle-aged', 'senior'], right=False)
    
    X = X.drop(columns=['age'])
    return X
```

```python
# Creación del FunctionTransformer
feature_extractor = FunctionTransformer(Feature_Extractor)
```
