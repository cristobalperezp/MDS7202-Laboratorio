### 3.3 Datos Nulos y Tratamiento

```python
from sklearn.impute import SimpleImputer
def data_cleaner(X):
    # Outliers: se tratarán sólo los valores negativos, y en age, se hará un tratamiento especial de los extremos, respecto de las demás, se utilizará 
    # un RobustScaler para tratarlos más adelante, ya que en son valores que podrían darse.

    # age: valores negativos y extremos
    # se asume que los valores negativos son errores de tipeo, por lo que se los convierte a positivos
    X.loc[:,'age'] = X.loc[:,'age'].apply(lambda x: -x if x<0 else x)
    # se asume que los valores extremos son errores de tipeo, por lo que se obtiene el residuo de la división por 100
    X.loc[:,'age'] = X.loc[:,'age'].apply(lambda x: x%100 if x>100 else x)
    # se vuelve a valor entero
    X.loc[:,'age'] = pd.to_numeric(X.loc[:,'age'], errors='coerce', downcast='integer')
    # negativos
    negative = ['num_bank_accounts','num_of_loan','delay_from_due_date','num_of_delayed_payment','changed_credit_limit','monthly_balance']
    for col in negative:
        X.loc[:,col] = X.loc[:,col].apply(lambda x: -x if x<0 else x)

    # occupation = '_______'
    X.loc[:,'occupation'] = X.loc[:,'occupation'].apply(lambda x: 'NoInfo' if x=='_______' else x)

    # payment_behaviour = '!@9#%8' 
    X.loc[:,'payment_behaviour'] = X.loc[:,'payment_behaviour'].apply(lambda x: 'NoInfo' if x=='!@9#%8' else x)


    # Nulos:
    nulls = ['num_of_delayed_payment','changed_credit_limit','num_credit_inquiries','credit_history_age','amount_invested_monthly','monthly_balance']
    imputer = SimpleImputer(strategy='mean')
    X[nulls] = imputer.fit_transform(X[nulls])
    return X
```

```python
# Tratamiento de datos
cleaner = FunctionTransformer(data_cleaner)
```

## 5 Optimización del Modelo

```python
# explicitly require this experimental feature
from sklearn.experimental import enable_halving_search_cv # noqa
# now you can import normally from model_selection
from sklearn.model_selection import HalvingGridSearchCV
```

```python
params = [
    # grid LGMB
    {
        "model": [LGBMClassifier(random_state=seed,n_jobs=-1)],
        'model__n_estimators': [100, 1000, 5000],
        'model__max_depth': [2, 3],
        'model__learning_rate': [1e-3, 1e-2, 1e-1]
    },
    # grid Random Forest
    {
        "model": [RandomForestClassifier(random_state=seed,n_jobs=-1)],
        'model__n_estimators': [100, 1000, 5000],
        'model__max_depth': [2, 3],
        'model__ccp_alpha': [0.001, 0.1],
    },


]
```

```python
from sklearn.feature_selection import SelectPercentile, f_classif
```

```python
clf_base = Pipeline(steps=[
                    ('cleaner', cleaner),
                    ("preprocessing", ct),
                    ('selection', SelectPercentile(f_classif, percentile=90)),
                    ("model", DummyClassifier(strategy='stratified',random_state=seed) )
])
```

```python
search = HalvingGridSearchCV(
    estimator=clf_base,
    param_grid=params,
    verbose=10,
    scoring='precision',
    cv=3,
    error_score='raise',
    n_jobs=-1,
    random_state=seed
)

search.fit(X_train,y_train)
```

```python
search.best_estimator_
```

```python
# Predict
y_pred = search.best_estimator_.predict(X_test)

# Resultados
print("Matriz de confusión\n", confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
print("\nPrecision-Score", precision_score(y_pred, y_test).round(2))
```
