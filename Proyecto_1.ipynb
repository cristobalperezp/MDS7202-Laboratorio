El origen de los duplicados puede ser haber utilizado un "outer join", que al unir dos dataframes con este metodo puede generar duplicados y valores nulos:

Un "outer join" combina registros de dos dataframes basándose en columnas comunes.
Hay tres posibles resultados en una "outer join": coincidencias, no coincidencias en el dataframe izquierdo y no coincidencias en el dataframe derecho.
Los duplicados pueden ocurrir cuando hay registros con múltiples coincidencias en la columna de unión, lo que genera combinaciones separadas en el resultado de la unión.
Los valores nulos se generan cuando hay registros que no tienen coincidencias en uno de los dataframes, lo que resulta en valores nulos en las columnas correspondientes del otro dataframe en el resultado de la unión.
Es fundamental tener en cuenta estos posibles resultados al usar un "outer join" para evitar problemas y asegurarse de obtener el resultado deseado. Una solución sencilla es eliminar los duplicados, quedandose solo con un dato del par.

## 4. Preprocesamiento

### 4.1 Declarar `ColumnTransformer`

*Esta sección consiste en generar los distintos pasos para preparar sus datos con el fin de luego poder crear su modelo.*

Dentro de los aspectos minimos que se les solicitarán en este desarrollo estan:

- Definir preprocesadores para datos categóricos y ordinales.
- Setear las transformaciones en un `ColumnTransformer`.
- Transformar todo el dataset.

**Notas**: 

- Las variables `id` y `fraud_bool` no deben ser transformadas, pero si pasadas a la siguiente etapa. (Estudiar la transformación `passthrough`).
- La salida de la transformación debe ser un pandas dataframe. Para esto, investiguen acerca de la nueva [`set_output` API](https://scikit-learn.org/stable/auto_examples/miscellaneous/plot_set_output.html). Si se levanta un error por matriz sparse, desactiven su uso en la transformación correspondiente.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler 
from sklearn.pipeline import Pipeline
from sklearn import set_config

# Configurar transform_output para que el resultado sea un DataFrame
set_config(transform_output='pandas')

# obtener nombre de todas las columnas numericas
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns.drop('id')

# obtener nombres de todas las columnas categoricas
categorical_columns = df.select_dtypes(include=['category']).columns.drop(['fraud_bool','month'])

# columnas ordinales
ordinal_columns = ['month']

# ColumnTransformer 
preprocessor = ColumnTransformer(
    [
    ('OneHot', OneHotEncoder(sparse=False, handle_unknown='ignore',drop='if_binary'), categorical_columns),
    ('Ordinal', OrdinalEncoder(handle_unknown='ignore'), ordinal_columns),
    ('Scaler', MinMaxScaler(), numeric_columns)
    ],
    remainder="passthrough"
    )
```

```python
pipe_pp = Pipeline([("Preprocesamiento", preprocessor)])
```

### 4.2 Transformar datos

```python
# Transformar todo el dataset
df_transformed = pipe_pp.fit_transform(df)

df_transformed
```
